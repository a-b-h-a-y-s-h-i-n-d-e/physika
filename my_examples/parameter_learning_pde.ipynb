```python
import torch
```

```python
def heat_equation(T, dx, alpha):
    nx = T.shape[0]
    f = torch.zeros(nx)

    f[1:-1] = alpha/dx ** 2 * (T[:-2] - 2 *T[1:-1] + T[2:])
    f[0] = 0
    f[-1] = 0
    return f

def forward_solver(alpha, T0, dx, dt, nt):
    T = T0
    for _ in range(nt):
        print("inside loop")
        result = heat_equation(T, dx, alpha)
        print(result)
        T = T + dt * heat_equation(T, dx, alpha)
    return T
```

```python


lx = 1.0
nx = 21
dx = lx / (nx - 1)
x = torch.linspace(0., lx, nx)

true_alpha = torch.tensor(0.4)
fourier = 0.49
dt = fourier * dx**2 / true_alpha
nt = 100

T0 = torch.sin(2 * torch.tensor(3.1459) * x)
```

```python
print(T0)
```

```python
nt = 1
forward_solver(true_alpha, T0, dx, dt, nt)
```

```python
u_true = forward_solver(true_alpha, T0, dx, dt, nt)
```

```python
def loss(alpha):
    u_pred = forward_solver(alpha, T0, dx, dt, nt)
    return torch.sum((u_pred - u_true)**2) / nx
```

```python
alpha = torch.tensor(0.1, requires_grad=True)
lr = 0.01
epochs = 1000
```

```python
from tqdm import tqdm

for i in tqdm(range(epochs)):
    l = loss(alpha)
    l.backward()
    with torch.no_grad():
        alpha -= lr * alpha.grad
        alpha.grad.zero_()
```
